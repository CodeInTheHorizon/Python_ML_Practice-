```python
from scipy.stats import skew
dataset = [10,25,14,26,35,45,67,90,40,50,60,10,16,18,20]

#calculate skewness
print(skew(dataset, axis=0, bias=True))
```

```python
# finding kurtosis
from scipy.stats import kurtosis

dataset =  [10,25,14,26,35,45,67,90,40,50,60,10,16,18,20]

print(kurtosis(dataset, axis=0, bias=True))
```

```python
# correlation
import numpy as np

x= np.array([10,25,14,26,35,45,67,90,40,50,60,10,16,18,20])
y= np.array([10,20,45,50,32])

def pearson_correlation(x,y):
  if len(x) == len(y):
    sum_XY = sum((x-x.mean())*(y-y.mean()))
    sum_X_squared = sum((x-x.mean())**2)
    sum_y_squared = sum((y-y.mean())**2)
    corr = sum_XY/np.sqrt(sum_X_squared*sum_y_squared)
    return corr

print(pearson_correlation(x,y))
print(pearson_correlation(x,x))
```

```python
#Visualization part

import pandas as pd
import seaborn as sns

data = {'name':['kshitij','aman','dhruv','nemish'], 'age':[20,20,18,20] }

df= pd.DataFrame(data)

#lineplot

sns.lineplot(x=df.index, y='age', data =df)
```
